# Calculate Tokens from ecxel file 

```python
# Import 
import pandas as pd
import tiktoken
```

```python
# Load Excel 
df = pd.read_excel("/Users/Buas/Desktop/Data Start/GitHub/Ano 2/2024-25c-fai2-adsai-group-group16/Data/transcription_whisper.xlsx")

# Initialize tokenizer
encoder = tiktoken.get_encoding("cl100k_base") 

# Count tokens
def count_tokens(text):
    return len(encoder.encode(text)) if isinstance(text, str) else 0

# Apply token count to each row
df["Token_Count"] = df["Text"].apply(count_tokens)

# Calculate the total number of tokens
total_tokens = df["Token_Count"].sum()

# Append total sum
summary_row = pd.DataFrame({"Text": ["TOTAL"], "Token_Count": [total_tokens]})
df = pd.concat([df, summary_row], ignore_index=True)

# Save to a new Excel file
df.to_excel("/Users/Buas/Desktop/Data Start/GitHub/Ano 2/2024-25c-fai2-adsai-group-group16/Data/transcribed_data_whisper_tokens.xlsx", index=False)
```
