```python
from pytubefix import YouTube
import assemblyai as aai
import os
import pandas as pd
```

# Step 1: Create the mp3 file from the youtube audio

```python
# as we croped the video we downlowded directly from youtube and removed the firrst 10 mints and after the 50

# We will not do this
# url input from youtube
yt = YouTube("https://www.youtube.com/watch?v=dN3mv5QiZLY&t=952s")

# extract only audio
video = yt.streams.filter(only_audio=True).first()

# set destination to save file
destination = ("/Users/celinewu/Documents/GitHub/2024-25c-fai2-adsai-group-group16/data")

# download the file
out_file = video.download(output_path=destination)

# save the file
base, ext = os.path.splitext(out_file)
new_file = base + '.mp3'
os.rename(out_file, new_file)
```

# Step 2: Extract the sentences

```python
aai.settings.api_key = "Enter API key here"
 
 
# Or use a publicly-accessible URL:
audio_file = (
    "/Users/celinewu/Documents/GitHub/2024-25c-fai2-adsai-group-group16/Data/La isla de las tentaciones Temporada 7 capitulo 7.mp3"
)
 
config = aai.TranscriptionConfig(language_code="es")
 
transcriber = aai.Transcriber(config=config)
 
transcript = transcriber.transcribe(audio_file)
 
if transcript.status == aai.TranscriptStatus.error:
    print(f"Transcription failed: {transcript.error}")
    exit(1)
 
sentences = transcript.get_sentences()
for sentence in sentences:
  print(sentence.text)
```

```python
# Limiting to first 300 sentences
limited_sentences = sentences[:300]

# Creating DataFrame
data = {'Sentence Number': list(range(1, len(limited_sentences) + 1)),
        'Text': [sentence.text for sentence in limited_sentences]}

df = pd.DataFrame(data)

# Saving to Excel
output_file = "/Users/celinewu/Documents/GitHub/2024-25c-fai2-adsai-group-group16/Data/Transcript_First_300_Sentences.xlsx"
df.to_excel(output_file, index=False)

print(f"First 300 sentences saved to {output_file}")
```
