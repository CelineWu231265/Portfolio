```python
import os
import shutil

def move_val_files(base_folder):
    # Define the subfolder names
    subfolders = ['images', 'masks']
    train_subfolder = 'train'
    test_subfolder = 'test'

    for subfolder in subfolders:
        # Paths to the train and test folders
        train_path = os.path.join(base_folder, subfolder, train_subfolder)
        test_path = os.path.join(base_folder, subfolder, test_subfolder)

        # Ensure the test folder exists
        os.makedirs(test_path, exist_ok=True)

        # Iterate through all files in the train folder
        for filename in os.listdir(train_path):
            # Check if "val" is in the file name
            if "val" in filename:
                # Construct full paths
                source_file = os.path.join(train_path, filename)
                destination_file = os.path.join(test_path, filename)

                # Move the file to the test folder
                shutil.move(source_file, destination_file)
                print(f"Moved: {source_file} -> {destination_file}")

# Set your base folder path here
base_folder = "/Users/celinewu/Documents/Processed_2024DS"

# Call the function
move_val_files(base_folder)
```
